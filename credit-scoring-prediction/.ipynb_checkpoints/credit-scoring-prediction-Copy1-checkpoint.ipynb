### Tratamiento de nulos, eliminar nulos

```python

print("\nCantidad total de datos nulos antes de eliminar:", df_banco.isnull().sum().sum())

# Eliminar los registros con valores nulos
df_banco = df_banco.dropna()

# Mostrar la cantidad total de datos nulos después de eliminar
print("Cantidad total de datos nulos después de eliminar:", df_banco.isnull().sum().sum())
```

```python
# Diccionario de mapeo
mapeo = {
    'default': {0: 'Good customer', 1: 'Bad customer'},
    'account_check_status': {'< 0 DM': 1, '0 <= ... < 200 DM': 2, '>= 200 DM / salary assignments for at least 1 year': 3, 'no checking account': 4},
    'duration_in_month': {1: '1 <= ... < 12', 2: '12 <= ... < 24', 3: '24 <= ... < 36', 4: '36 <= ... < 48', 5: '48 <= ... < 60', 6: '60 <= ... < 72'},
    'credit_history': {'no credits taken/all credits paid back duly': 1, 'all credits at this bank paid back duly': 2, 'existing credits paid back duly till now': 3, 'delay in paying off in the past': 4, 'critical account/other credits existing (not at this bank)': 5},
    # Agrega más columnas según sea necesario
}

# Aplicar el mapeo a cada columna categórica
for columna, mapeo_columna in mapeo.items():
    df_banco[columna] = df_banco[columna].map(mapeo_columna)

# Mostrar el resultado
print(df_banco.head())
```
